```python

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

file_path='dataset.csv'
df = pd.read_csv(file_path)
```

```python
df.head()
```

```python
df.describe()
df.dropna()
```

We add label encoder to convert object datatype to int64.

```python
from sklearn.preprocessing import LabelEncoder

le=LabelEncoder()

for col in df.columns:
    if df[col].dtype=='object':
        df[col]=le.fit_transform(df[col])
```

```python
print(df.shape)
print(df.columns)
```

We define hyperparameters for the model.

```python
x = df[[' no_of_dependents', ' education', ' self_employed', ' income_annum', ' loan_amount', ' loan_term', ' cibil_score', ' residential_assets_value', ' commercial_assets_value',' luxury_assets_value', ' bank_asset_value']]
y = df[' loan_status']
```

We import Train-Test Split from sklearn library.


```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(x,y,test_size = 0.30, random_state = 0)
X_train = X_train.replace([np.inf, -np.inf], np.nan)
X_train = X_train.dropna()
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy = 'mean')
X_train_imp = imputer.fit_transform(X_train)
X_test_imp = imputer.fit_transform(X_test)
Y_train_imp = imputer.fit_transform(Y_train.values.reshape(-1,1))
```

```python
from sklearn.metrics import accuracy_score,recall_score,precision_score,f1_score,confusion_matrix
```

```python
from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier()
model.fit(X_train_imp, Y_train_imp)

Y_predict = model.predict(X_test_imp)

accuracy = accuracy_score(Y_predict, Y_test)*100
recall = recall_score(Y_predict, Y_test)*100
f1 = f1_score(Y_predict, Y_test)*100
precision = precision_score(Y_predict, Y_test)*100
matrix = confusion_matrix(Y_predict, Y_test)
print("accuracy",accuracy, '%')
print("recall",recall, '%')
print("precision",precision, '%')
print("f1",f1, '%')
print("confusion matrix",matrix,sep="\n")

```
